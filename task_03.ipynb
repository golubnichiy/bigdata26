#### Практическая работа – Поиск элементов по шаблонам

1. Найти все слова, начинающиеся с большой буквы
2. Найти все слова из 4 букв
3. Найти все слова, содержащие только буквы
4. Найти все числа с плавающей точкой
5. Извлечь все доменные имена
6. Найти все ссылки
7. Найти все слова, где есть две одинаковые буквы подряд
8. Найти повторяющиеся слова подряд

```python
text = "Это текст содержит очень много слов в том числе С Большой Буквы и с мылых букв А4, А3. Формат записи 3.14 и 2,33  .5 Сайт site.ru и ya.com www.site.ru. http://1999.ru ветренный опять опять"
```

```python
import re
```

```python
re.findall(
    r"\b([A-ZА-ЯЁ]\w+)\b", 
    text
)
```

```python
re.findall(
    r"\b\w{4}\b", 
    text
)
```

```python
re.findall(
    r"\b[A-zА-яёЁ]+\b", 
    text
)
```

```python
re.findall(
    r"\b\d+[,\.]\d+\b", 
    text
)
```

```python
re.findall(
    r"\b\w+(?:\.\w+)+\b", 
    text
)
```

```python
re.findall(
    r"\b\w{2,}(?:\.\w+)*\.[A-zА-я]+\b", 
    text
)
```

```python
re.findall(
    r"\b\w+://\w{2,}(?:\.\w+)*\.[A-zА-я]+(?:/\S*)?\b", 
    text
)
```

```python
[i[0] for i in re.findall(
    r"\b(\w*([A-zА-яёЁ])\2\w*)\b",
    text
)]
```

```python
re.findall(
    r"\b(\w+)\s\1\b", 
    text
)
```

```python
[i[0] for i in re.findall(
    r"\b((\w+)\s\2)\b",
    text
)]
```
