#### Практическая работа – Создание шаблонов

1. Создайте шаблон для проверки электронной почты
2. Создайте шаблон для проверки простых паролей (менее 6 символов, только цифры, нет спецсимволов)
3. Создайте шаблон для проверки телефонных номеров (+7-999-999-99-99, +7(999)999-99-99, +7(999)999-9999, +7-999-999-9999, 8-999-999-99-99)
4. Создайте шаблон для проверки ФИО (Фамилия Имя Отчество, Фамилия И.О.)

```python
import re
reg_1 = r"([A-Za-z]{2,}@[A-Za-z]{2,}(\.[A-Za-z]{2,})+|[А-Яа-я]{2,}@[А-Яа-я]{2,}(\.[А-Яа-я]{2,})+)"
st = 'Моя электронная почта artem@mail.ru'
print(re.search(reg_1, st))
```

```python
reg_2 = r"(^.{1,6}$)|(^\d+$)|(^\w+$)"
st = "1234567s!"
print(re.match(reg_2, st))
```

```python
reg_3 = r"^(\+7|8)((\(\d{3}\))|(-\d{3}-)|(\d{3}))\d{3}-?\d{2}-?\d{2}"
st = "8(999)999-9999"
print(re.match(reg_3, st))
```

```python
reg_4 = r"([А-ЯЁ][а-яё]+ [А-ЯЁ][а-яё]+ [А-ЯЁ][а-яё]+)|([А-ЯЁ][а-яё]+ [А-ЯЁ]\. ?[А-ЯЁ]\.)"
st = "Иванов А. И."
print(re.match(reg_4, st))
```
